# **1. QWen generative Large Language Model Wanx2.1 in generating images from text**


```python
from http import HTTPStatus
from urllib.parse import urlparse, unquote
from pathlib import PurePosixPath
import requests
from dashscope import ImageSynthesis
import os
from IPython.display import Image

prompt = "A flower shop with exquisite windows, a beautiful wooden door, and flowers on display."


print('----sync call, please wait a moment----')
rsp = ImageSynthesis.call(api_key="sk-73009b1ebbb84cc1aa2b3efeba348f87",
                          model="wanx2.1-t2i-turbo",
                          prompt=prompt,
                          n=1,
                          size='1024*1024')
print('response: %s' % rsp)
if rsp.status_code == HTTPStatus.OK:
    # Save figure under current folder
    for result in rsp.output.results:
        file_name = PurePosixPath(unquote(urlparse(result.url).path)).parts[-1]
        with open('./%s' % file_name, 'wb+') as f:
            f.write(requests.get(result.url).content)
else:
    print('sync_call Failed, status_code: %s, code: %s, message: %s' %
          (rsp.status_code, rsp.code, rsp.message))
image_path = file_name
Image(image_path)
```

# **2. QWen generative Large Language Model Wanx2.1 in generating videos from text**
Please save the video once generated by the LLM model since it will soon be deleted from the Ali cloud platform

```python
from http import HTTPStatus
# dashscope sdk >= 1.22.1
from dashscope import VideoSynthesis
from IPython.display import Video
from http import HTTPStatus
from urllib.parse import urlparse, unquote
from pathlib import PurePosixPath
import requests
from dashscope import ImageSynthesis
import os
from IPython.display import Image
# call sync api, will return the result
print('please wait...')
rsp = VideoSynthesis.call(model='wanx2.1-t2v-turbo',
                          api_key="sk-73009b1ebbb84cc1aa2b3efeba348f87",
                          prompt='workers are constructing a modern modular high rise building',
                          size='1280*720')
print(rsp)
if rsp.status_code == HTTPStatus.OK:
    file_name = PurePosixPath(unquote(urlparse(rsp.output.video_url).path)).parts[-1]
    file_path = f'./{file_name}'
    with open('./%s' % file_name, 'wb+') as f:
        f.write(requests.get(rsp.output.video_url).content)
    Video(file_path)
else:
    print('Failed, status_code: %s, code: %s, message: %s' %
          (rsp.status_code, rsp.code, rsp.message))
Video(file_path)
```
