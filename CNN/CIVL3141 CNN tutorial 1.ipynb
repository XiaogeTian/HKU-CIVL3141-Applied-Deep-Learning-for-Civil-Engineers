# CNN background introduction
Convolutional Neural Networks (CNNs) are inspired by human visual effects, which are the layered architecture of cortex. Just as the figure below, the visual cortex and CNN framework are displayed with part **a** and **b**. ## 1. Environment setup

```python
import os
import cv2
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from sklearn.model_selection import train_test_split
import random
```

## 2. Dataset preparation

```python
# Defining dataset 
class CrackDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # transfer to RBG format
        img = cv2.resize(img, (64, 64))  # adjust the resolution of img

        if self.transform:
            img = self.transform(img)

        return img, label

# path of cracks
positive_dir = "../data/crack/Positive"
negative_dir = "../data/crack/Negative"

# obtaining labels and images
positive_paths = [os.path.join(positive_dir, f) for f in os.listdir(positive_dir) if f.endswith('.jpg')]
negative_paths = [os.path.join(negative_dir, f) for f in os.listdir(negative_dir) if f.endswith('.jpg')]


num_samples_per_class = 500
positive_paths = random.sample(positive_paths, min(len(positive_paths), num_samples_per_class))
negative_paths = random.sample(negative_paths, min(len(negative_paths), num_samples_per_class))
image_paths = positive_paths + negative_paths

labels = [1] * len(positive_paths) + [0] * len(negative_paths)

    
# data prepocessed
transform = transforms.Compose([
    transforms.ToTensor(),  # normalized to [0,1]
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))  # normalized to [-1,1]
])

# seperating training set and testing set
train_paths, test_paths, train_labels, test_labels = train_test_split(image_paths, labels, test_size=0.2, random_state=42)

# creating dataset
train_dataset = CrackDataset(train_paths, train_labels, transform=transform)
test_dataset = CrackDataset(test_paths, test_labels, transform=transform)

# creating DataLoader
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print(f"training set size: {len(train_dataset)}")
print(f"testing set size: {len(test_dataset)}")
```

## 3. CNN model establishment

```python
import torch.nn as nn
import torch.nn.functional as F

class CrackCNN(nn.Module):
    def __init__(self):
        super(CrackCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, 3)  # input 3 channels，output 32 channels，convolutional 3x3
        self.pool = nn.MaxPool2d(2, 2)    # 2x2 pooling
        self.conv2 = nn.Conv2d(32, 64, 3)
        self.conv3 = nn.Conv2d(64, 64, 3)
        self.fc1 = nn.Linear(64 * 6 * 6, 128)  # fully connected neural network
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, 1)  # output the final classification result

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(-1, 64 * 6 * 6)  # flatting
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = torch.sigmoid(self.fc2(x))  # classification rate
        return x

# Initializing model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CrackCNN().to(device)
print(model)
```

## 4. Loss function, optimizer, and learning rate setup

```python
import torch.optim as optim

criterion = nn.BCELoss()  # defining loss function
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

## 5. Training loop

```python
num_epochs = 10

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device).float()

        # forward process
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs.squeeze(), labels)

        # backward process and optimizing traininable weights
        loss.backward()
        optimizer.step()

        # calculation
        running_loss += loss.item()
        predicted = (outputs.squeeze() > 0.5).float()
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    epoch_loss = running_loss / len(train_loader)
    epoch_acc = 100 * correct / total
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.2f}%")
```

## 6. Results visualization

```python
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device).float()
        outputs = model(images)
        predicted = (outputs.squeeze() > 0.5).float()
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

test_acc = 100 * correct / total
print(f"test set accuracy: {test_acc:.2f}%")
```

```python
import matplotlib.pyplot as plt
# classify test set and visualize the results
def classify_and_visualize(test_loader, model, device, num_images):
    model.eval()
    images_shown = 0
    fig, axes = plt.subplots(5, 5, figsize=(8, 8))  # display the results
    axes = axes.flatten()

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device).float()
            outputs = model(images)
            predicted = (outputs.squeeze() > 0.5).float()

            for i in range(images.size(0)):
                if images_shown >= num_images:
                    break

                # transpose the image data to figures
                img = images[i].cpu().numpy().transpose(1, 2, 0)  # transpose from CHW to HWC
                img = img * 0.5 + 0.5  # transpose to [0,1]
                img = np.clip(img, 0, 1)

                # true labels and predicted labels
                true_label = "cracked" if labels[i].item() == 1 else "no cracks"
                pred_label = "cracked" if predicted[i].item() == 1 else "no cracks"

                # display figures and labels
                axes[images_shown].imshow(img)
                axes[images_shown].set_title(f"true: {true_label}\n pred: {pred_label}")
                axes[images_shown].axis('off')
                images_shown += 1

            if images_shown >= num_images:
                break

    plt.tight_layout()
    plt.show()

# test and visualization
classify_and_visualize(test_loader, model, device, num_images=25)
```
